# Pregunta 3
Para esta parte se requieren los mejores 5 animes, lo que se determina por los mejores Scores de estos. Primero ordenamos los animes en orden decreciente de Score aplicando la funcion sort_values a nuestro dataset data_animes con argumento 'Score' y ascending=False para el orden de mayor a menor. A esta ordenacion le llamamos top_5 y obtuvimos las primeras 5 filas con .head(5).

```python
by_score = data_animes.sort_values('Score',ascending=False)
top_5 = by_score.head(5)
```

Se obtienen el nombre, la cantidad de Completed, Watching y Plan to Watch con top_5['Columna correspondiente']

```python
name_top_5 = top_5['Name']
completed_top_5 = top_5['Completed']
watching_top_5 = top_5['Watching']
plantowatch_top_5 = top_5['Plan to Watch']
```

En esta seccion definimos el grafico con su nombre para darle un tamaño adecuado. Luego convertimos los nombres obtenidos anteriormente a lista usando tolist() para darle nombre a las variables en el grafico. 
Para graficar le damos labels en la lista de nombres y usamos plt.pie de los datos convertidos a lista corresponidentes a cada uno. Además les definimos un titulo y un eje con plt.title y plt.axis respectivamente.

```python
pie_chart_completed = plt.figure(figsize=(10,8))
tasks = completed_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('Top 5 Completed Comparison')
plt.axis('equal')
plt.show()
```

De este grafico tenemos que Fullmetal Alchemist: Brotherhood es el que mas ha sido completado de los top 5. Shingeki no Kyojin: The Final Season tiene un valor de 0% debido a su baja cantidad con respecto a los otros.

```python
pie_chart_watching = plt.figure(figsize=(10,8))
tasks = watching_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('Top 5 Watching Comparison')
plt.axis('equal')
plt.show()
```

Tenemos que el 50% de los mejores 5 en la categoria Watching corresponde a Shingeki no Kyojin: The Final Season

```python
pie_chart_plantowatch = plt.figure(figsize=(10,8))
tasks = plantowatch_top_5.tolist()

my_labels = name_top_5.tolist()
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('Top 5 Plan to Watch Comparison')
plt.axis('equal')
plt.show()
```

De los top 5 en la categoria Plan to Watch, un 32,5% corresponde a Steins;Gate

Para hacer el grafico de torta de las tres categorias de los top 5, sumamos los elementos de cada set de datos con sum() y los guardamos en una variable con nombre para reconocerlos facilmente. Esto lo graficamos entregandolo a la funcion plt.pie en forma de lista y dandole etiquetas, titulo y eje de la misma forma ya realizada.

```python
total_completed = completed_top_5.sum()
total_watching = watching_top_5.sum()
total_plantowatch = plantowatch_top_5.sum()
```

```python
pie_chart_totales = plt.figure(figsize=(10,8))
tasks = [total_completed, total_watching, total_plantowatch]

my_labels = ['Completed', 'Watching', 'Plan to Watch']
plt.pie(tasks,labels=my_labels,autopct='%1.1f%%')
plt.title('Anime Aggregated')
plt.axis('equal')
plt.show()
```

De los mejores 5 animes, un 62,4% pertenecen a la categoria Completed, un 20,2% a la categoria Plan to Watch y un 17,4% a la categoria Watching.

Tomamos la columna 'Year' y lo pasamos a floats con to_numeric, junto con errors='coerce' considerando que pueden existir datos que no son numeros. En el paso siguiente definimos dat como los animes desde el año 2000 en adelante especificando data_animes['Year'] >= 2000. Esto ultimo se ordena de menor a mayor, se filtran por año y se calcula el promedio con mean().
Para graficar, usamos regplot de seaborn, a esta le entregamos en data los datos a graficar y lo que corresponde al eje Y.

```python
data_animes['Year'] = pd.to_numeric(data_animes['Year'], errors='coerce')
dat = data_animes[data_animes['Year'] >= 2000]

by_year = dat.sort_values('Year',ascending=True)
grupo=by_year.groupby("Year").mean()

fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Score", ax=ax, order=3, ci=None).set_title("Score Means by Year")
plt.xticks(rotation=45)
plt.show()
```

En este grafico de promedio y regresion no lineal (orden 3) sirve para poder graficar los scores promedio por año y además se representa la regresion como la relación entre los scores y años de cada uno. Además nos dice como se comportan los scores a lo largo de los años

```python
fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Watching", ax=ax, order=3, ci=None).set_title("Watching Means by Year")
plt.xticks(rotation=45)
plt.show()
```

En este grafico podemos ver que durante los primeros 15 años se mantiene un promedio muy bajo por año de la gente que esta viendo la serie, para luego, en los ultimos 5 años subir el promedio hasta 50000 aproximadamente

```python
fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Completed", ax=ax, order=3, ci=None).set_title("Completed Means by Year")
plt.xticks(rotation=45)
plt.show()
```

El peak del promedio de personas que ya completaron la serie se alcanza entre los años 2012 y 2015 para luego caer hasta casi 0. La regresion en el peak se dispersa un tanto del promedio.

```python
fig, ax= plt.subplots()
sns.regplot(data=grupo,x=grupo.index[True],y="Plan to Watch", ax=ax, order=3, ci=None).set_title("Plan to Watch Means by Year")
plt.xticks(rotation=45)
plt.show()
```

En este grafico de regresion no lineal se ve un incremento del promdio a medida que pasan los años de la gente que tiene planeado ver la serie.
